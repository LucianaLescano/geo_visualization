Para ense침ar este ejemplo, utilizamos un dataset donde tenemos ya unas columnas de latitud y longitud 

```python
df = pd.read_csv('data/bici_clean.csv')
```

```python
df.head()
```

```python
#Como hay muchos datos y estamos haciendo pruebas, vamos a quedarnos solo con los de 2019
df_2019 = df[(df.year == 2019)]
```

```python
#쮺u치ntas clasificaciones de festividad hay?

df_2019.festividad.unique()
```

```python
lat2=40.428990
lon2=-3.681209
```

```python
loc_2 = [lat2,lon2]
```

```python
map_2 = folium.Map(location=loc_2, zoom_start=15)
#map_2
```

Vamos a iterar sobre todas las filas del dataframe buscando qu칠 hay dentro de cada fila en la columna. 
Pondremos un icono a los accidentes en funci칩n de la columna "festivo"
Pero tendremos que guardar las configuraciones y a침adir el Marker al final y todo esto dentro de un bucle 游뱘游뱚
- Primero la config del tooltip
- Despu칠s el iconito en funci칩n del festivo
- Por 칰ltimo a침adir cada marker

Keep Calm, lo vemos 游뗶

```python
#Bueno, recordad que lo primerito es crear un mapa base al que a침adimos cosas
# map_2 = Map(location=[40.428990,-3.681209],zoom_start=15)
```

```python
for i,row in df_2019.iterrows():
    #popup distrito
    distrito = {
        'location':[row['lat'], row['lon']],
        'tooltip': row['distrito']
    }
    if row['festividad'] == 'Festivo':
        icon=Icon(color='green',
            prefix='fa',
            icon='glass',
            icon_color='black')
    elif row['festividad'] == 'Vispera':
        icon=Icon(color='blue',
            prefix='fa',
            icon='home',
            icon_color='black')
    else:
        icon=Icon(color='red',
            prefix='fa',
            icon='briefcase',
            icon_color='black')
    Marker(**distrito, icon=icono).add_to(map_2)
map_2
```

### Mapa de Calor

Vamos a comprobar cu치ntas categor칤as hay en la columna de horario, para hacer un mapa de calor. Primero hacemos un mapa con el horario de ma침ana, y despu칠s a침adimos el horario de tarde con el Layer Control. 
All치 vamos 游댠

```python
df.horario.unique()
```

```python
map_3 = Map(location=[40.428990,-3.681209],zoom_start=15)
```

```python
morning = df[df.horario=='Ma침ana']
morning_group = folium.FeatureGroup(name='Ma침ana')
HeatMap(data=morning[['lat', 'lon']], radius=15).add_to(morning_group)
morning_group.add_to(map_3)
map_3
```

```python
afternoon = df[df.horario=='Tarde']
afternoon_group = folium.FeatureGroup(name='Tarde')
HeatMap(data=afternoon[['lat', 'lon']], radius=15).add_to(afternoon_group)
afternoon_group.add_to(map_3)
folium.LayerControl(collapsed=False).add_to(map_3)
map_3
```
